```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_values_2bodies_base = np.genfromtxt("closed orbits//planet_posiz.plot", usecols = 0)
y_values_2bodies_base = np.genfromtxt("closed orbits//planet_posiz.plot", usecols = 1)

x_values_2bodies_high = np.genfromtxt("closed orbits//planet_posiz_highervel.plot", usecols = 0)
y_values_2bodies_high = np.genfromtxt("closed orbits//planet_posiz_highervel.plot", usecols = 1)

x_values_2bodies_slow = np.genfromtxt("closed orbits//planet_posiz_smallervel.plot", usecols = 0)
y_values_2bodies_slow = np.genfromtxt("closed orbits//planet_posiz_smallervel.plot", usecols = 1)

timefile_values_base = np.genfromtxt("closed orbits//planet_time.plot", usecols = 0)
timefile_values_high = np.genfromtxt("closed orbits//planet_time_highervel.plot", usecols = 0)
timefile_values_slow = np.genfromtxt("closed orbits//planet_time_smallervel.plot", usecols = 0)

print(len(x_values_2bodies_base))
print(len(x_values_2bodies_high))
print(len(x_values_2bodies_slow))
print(len(timefile_values_base))
```

```python
x_planet, y_planet, x_sun, y_sun = [], [], [], []
x_planet_fast, y_planet_fast, x_sun_fast, y_sun_fast = [], [], [], []
x_planet_slow, y_planet_slow, x_sun_slow, y_sun_slow = [], [], [], []

time_base, time_fast, time_slow = [], [], []

#lettura del file e creazione array con dati
for i in range (0, len(x_values_2bodies_base)):
    if i%2 == 0:
        x_sun.append(x_values_2bodies_base[i])
        y_sun.append(y_values_2bodies_base[i])
        x_sun_fast.append(x_values_2bodies_high[i])
        y_sun_fast.append(y_values_2bodies_high[i])
        x_sun_slow.append(x_values_2bodies_slow[i])
        y_sun_slow.append(y_values_2bodies_slow[i])
    else: 
        x_planet.append(x_values_2bodies_base[i])
        y_planet.append(y_values_2bodies_base[i])
        x_planet_fast.append(x_values_2bodies_high[i])
        y_planet_fast.append(y_values_2bodies_high[i])
        x_planet_slow.append(x_values_2bodies_slow[i])
        y_planet_slow.append(y_values_2bodies_slow[i])

for i in range (0, int(len(timefile_values_base)/4)):
    time_placeholder = (4*i)+1
    time_base.append(timefile_values_base[time_placeholder])
    time_fast.append(timefile_values_high[time_placeholder])
    time_slow.append(timefile_values_slow[time_placeholder])
```

```python
fig, ax = plt.subplots(1,2, figsize = (20,10))
plt.rcParams.update({"figure.facecolor":"white", "font.size":15})
fig.suptitle("Bound orbits", fontsize=20)

ax[0].plot(x_sun, y_sun, color="r", marker = "o", markersize = "10", ls="", label="Sun")
ax[0].plot(x_planet[:90], y_planet[:90], color="b", marker = "o", markersize = "5", ls="", label="Planet")
ax[0].plot(x_planet_slow[:45], y_planet_slow[:45], color="g", marker = "o", markersize = "5", ls="", label="Planet (smaller initial speed)")
ax[0].plot(x_planet_fast, y_planet_fast, color="y", marker = "o", markersize = "5", ls="", label="Planet (higher initial speed)")
ax[0].grid(ls="--", linewidth=.5)
ax[0].set_title("Orbits", fontsize=17)
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

ax[1].plot(time_base, x_planet, color="b", marker = "", markersize = "5")
ax[1].plot(time_fast, x_planet_fast, color="y", marker = "", markersize = "5")
ax[1].plot(time_slow, x_planet_slow, color="g", marker = "", markersize = "5")
ax[1].set_title('"x" position as function of time')
ax[1].set_xlabel("time")
ax[1].set_ylabel("x position")
ax[1].grid(ls="--", linewidth=.5)

fig.legend(loc = "lower center", ncol = 4)
plt.show()
```
